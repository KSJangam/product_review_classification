```python
from google.colab import drive

drive.mount('/content/gdrive')
unlabeled_file = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/archive/unlabeled_150k.csv')
unlabeled_file['Text'][0]
```

```python
text=unlabeled_file['Text'][0]
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("KJangam/amznproducts")

inputs = tokenizer(text, return_tensors="tf")
```

```python
from transformers import TFAutoModelForSequenceClassification

model = TFAutoModelForSequenceClassification.from_pretrained("KJangam/amznproducts")

logits = model(**inputs).logits
```

```python
id2label={0: 'grocery gourmet food',
 1: 'toys games',
 2: 'baby products',
 3: 'pet supplies',
 4: 'beauty',
 5: 'health personal care'}
predicted_class_id = int(tf.math.argmax(logits, axis=-1)[0])

model.config.id2label[predicted_class_id]
```

```python
for num in range(20):
  print(unlabeled_file['Title'][num])
  print(unlabeled_file['Text'][num])
  inputs = tokenizer(unlabeled_file['Text'][num], return_tensors="tf")
  logits = model(**inputs).logits

  predicted_class_id = int(tf.math.argmax(logits, axis=-1)[0])

  print(model.config.id2label[predicted_class_id])
  print('---------------------------------------------------')
```
